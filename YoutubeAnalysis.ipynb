# Content Categories


```python
category_list = df['Categories'].str.split(',')

categories = [category for categories in category_list for category in categories]

category_counts = pd.Series(categories).value_counts()

category_counts_df = pd.DataFrame({'categories': category_counts.index, 'counts': category_counts.values})

category_counts_df
```

##### Top 10 Categories

```python
top_10_categories=category_counts_df['categories'].head(10)
top_10_categories
```

```python
plt.barh(top_10_categories,category_counts_df['counts'][:10])
plt.gca().invert_yaxis() 
plt.show()
```

# Category with exceptional performance matrix


```python
avg=df['Suscribers'].mean()
exceptional=df.groupby('Categories')['Suscribers'].mean().sort_values(ascending=False)
exceptional=exceptional[exceptional>avg]
exceptional.plot(kind='bar')
plt.ylabel('Average Subscribers')
plt.show()
```

# Benchmarking

##### Above average Subscribers

```python
avg_numeric=df[['Suscribers','Visits','Likes','Comments']].mean()
avg_numeric
```

```python
li=df[['Categories','Suscribers']]
avg_sub=li.groupby('Categories')['Suscribers'].mean().reset_index()
above_avg_sub=avg_sub[avg_sub['Suscribers']>avg_numeric['Suscribers']][['Categories','Suscribers']]
above_avg_sub
```

```python
plt.barh(above_avg_sub['Categories'],above_avg_sub['Suscribers'])
plt.xlabel('Subscribers')
plt.ylabel('Categories')
plt.show()
```

##### Above average visits

```python
li=df[['Categories','Visits']]
avg_visits=li.groupby('Categories')['Visits'].mean().reset_index()
above_avg_visits=avg_visits[avg_visits['Visits']>avg_numeric['Visits']][['Categories','Visits']]
above_avg_visits
```

```python
plt.barh(above_avg_visits['Categories'],above_avg_visits['Visits'])
plt.xlabel('Visits')
plt.ylabel('Categories')
plt.show()
```

##### Above average likes

```python
li=df[['Categories','Likes']]
avg_Likes=li.groupby('Categories')['Likes'].mean().reset_index()
above_avg_Likes=avg_Likes[avg_Likes['Likes']>avg_numeric['Likes']][['Categories','Likes']]
above_avg_Likes
```

```python
plt.barh(above_avg_Likes['Categories'],above_avg_Likes['Likes'])
plt.xlabel('Likes')
plt.ylabel('Categories')
plt.show()
```

```python
li=df[['Categories','Comments']]
avg_Comments=li.groupby('Categories')['Comments'].mean().reset_index()
above_avg_Comments=avg_Comments[avg_Comments['Comments']>avg_numeric['Comments']][['Categories','Comments']]
above_avg_Comments
```

```python
plt.barh(above_avg_Comments['Categories'],above_avg_Comments['Comments'])
plt.xlabel('Comments')
plt.ylabel('Categories')
plt.show()
```

##### Top Performers

```python
top_performers=df.groupby('Username')['Suscribers'].mean().sort_values(ascending=False).reset_index().head(5)
top_performers
```

```python
plt.pie(top_performers['Suscribers'],labels=top_performers['Username'],autopct='%1.1f%%')
plt.show()
```

# Recommendations 

```python
recommendations = df.groupby('Categories')['Categories'].count().sort_values(ascending=False).head(10)
recommendations = pd.DataFrame({'Category_Count': recommendations})
recommendations.index.name = 'Categories'
content_recommendations = recommendations.reset_index()
content_recommendations
```

```python
plt.barh(content_recommendations['Categories'],content_recommendations['Category_Count'])
plt.show()
```
